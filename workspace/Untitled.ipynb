```python
import os
import numpy as np

root = 'data'
# 遍歷 subdir 
features = []
for subdir in os.listdir(root):
    try:
        for f in os.listdir(os.path.join(root,subdir,'KEYPOINT_DIR')):
            feature = []
            dir_path = os.path.join(root,subdir,'KEYPOINT_DIR',f)
            for i in os.listdir(dir_path):
                if i.endswith('.npy'):
                    feature.append(np.load(os.path.join(dir_path, i)))
            feature = np.array(feature)
            features.append(feature)
    except:
        pass
```
